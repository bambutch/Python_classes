## Создание списка словарей

Используем известные нам литералы: `[]` - для списка, `{}` - для словаря.  
Для повышения читабельности кода - используем переносы.

Создадим список словарей на основе таблицы отгрузок, заголовки (ключи) заменим на более удобные в работе.

```python
# создание списка словарей

sales_list = [
    {
        'station': 'ст. Новоярославская', 
        'source':'ЯНОС', 
        'item': 'АИ-92',
        'contract':840, 
        'sale': 821.1, 
        'price': 37780
    },
    
    {
        'station': 'ст. Яничкино', 
        'source':'МНПЗ',
        'item': 'АИ-92',
        'contract':600, 
        'sale': 597.523, 
        'price': 37740
    },
    
    {
        'station': 'ст. Комбинатская', 
        'source':'ОНПЗ',
        'item': 'АИ-92',
        'contract':1320, 
        'sale': 1321.002, 
        'price': 36690
    }
]

# выведем список
sales_list
```

## Доступ к элементам словаря

Словарь - элемент списка, доступ к нему осуществляется по индексу.  
Доступ к значениям словаря осуществляется по ключу.

```python
# создание списка словарей

sales_list = [
    {
        'station': 'ст. Новоярославская', 
        'source':'ЯНОС', 
        'item': 'АИ-92',
        'contract':840, 
        'sale': 821.1, 
        'price': 37780
    },
    
    {
        'station': 'ст. Яничкино', 
        'source':'МНПЗ',
        'item': 'АИ-92',
        'contract':600, 
        'sale': 597.523, 
        'price': 37740
    },
    
    {
        'station': 'ст. Комбинатская', 
        'source':'ОНПЗ',
        'item': 'АИ-92',
        'contract':1320, 
        'sale': 1321.002, 
        'price': 36690
    }
]

# выведем отгрузку со ст. Яничкино и цену АИ-92
print(f'Данные по отгрузке: {sales_list[1]}')
print(f'Цена отгрузки АИ-92 со ст.Яничкино: {sales_list[1]["price"]}')
```

## Добавление словаря в список

Используем `append()`

```python
# создание списка словарей

sales_list = [
    {
        'station': 'ст. Новоярославская', 
        'source':'ЯНОС', 
        'item': 'АИ-92',
        'contract':840, 
        'sale': 821.1, 
        'price': 37780
    },
    
    {
        'station': 'ст. Яничкино', 
        'source':'МНПЗ',
        'item': 'АИ-92',
        'contract':600, 
        'sale': 597.523, 
        'price': 37740
    },
    
    {
        'station': 'ст. Комбинатская', 
        'source':'ОНПЗ',
        'item': 'АИ-92',
        'contract':1320, 
        'sale': 1321.002, 
        'price': 36690
    }
]


# дополнительный словарь
sales_added = {
    'station': 'ст. Стенькино II', 
    'source':'РНПК',
    'item': 'АИ-92',
    'contract':2400, 
    'sale': 2396.336, 
    'price': 37710
    }

# добавляем словарь в список словарей
sales_list.append(sales_added)

# выводим список словарей
sales_list
```

## Цикл for в списке словарей

Найдем суммарное количество отгруженного АИ-92. Для этого используем вложенный цикл for.

```python
# создание списка словарей

sales_list = [
    {
        'station': 'ст. Новоярославская', 
        'source':'ЯНОС', 
        'item': 'АИ-92',
        'contract':840, 
        'sale': 821.1, 
        'price': 37780
    },
    
    {
        'station': 'ст. Яничкино', 
        'source':'МНПЗ',
        'item': 'АИ-92',
        'contract':600, 
        'sale': 597.523, 
        'price': 37740
    },
    
    {
        'station': 'ст. Комбинатская', 
        'source':'ОНПЗ',
        'item': 'АИ-92',
        'contract':1320, 
        'sale': 1321.002, 
        'price': 36690
    }
]


# цикл for - общая отгрузка и выручка
total_sale = 0
total_revenue = 0
for item in sales_list:
    total_sale +=  item['sale']
    total_revenue += item['sale'] * item['price']
print(f'Суммарная отгрузка АИ-92: {total_sale} тонн')
print(f'Выручка: {total_revenue :.2f} рублей')
```
