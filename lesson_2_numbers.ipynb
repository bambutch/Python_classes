### Сложение

Оператор  `+`

Рассчитаем общий объем заключенных договоров и запишем результат в переменную `sale`.

image.png

```python
sale = 840 + 600
print(sale)
```

Операнды 840 и 600 являются целыми числами. Результат сложения - целое число. Проверим, используя функцию `type()`.

```python
print(type(sale))
```

Операнды могут быть числами разных типов - `int` и `float`.

```python
# 840.0 - тип float, 600 - тип int
sale = 840.0 + 600
print(sale)
```

Операнд 840.0 - вещественное число, 600 - целое число. Результат сложения - вещественное число.

```python
print(type(sale))
```

Сложение двух целых чисел всегда приводит к целочисленному результату **int**.  
**int + int ---> int**  

Сложение целого и вещественного числа дает в результате вещественное число **float**.  
**float + int ---> float**

```python
print(type(price_delta))
```

Операнды могут быть числами разных типов - `int` и `float`.

```python
# 37780.0 - тип float, 37740 - тип int
price_delta = 37780.0 - 37740
print(price_delta)
```

Операнд 37780.0 - вещественное число, 37740 - целое число. Результат вычитания - вещественное число.

```python
print(type(price_delta))
```

Вычитание двух целых чисел дает целочисленный результат **int**.  
**int - int ---> int**  

Вычитание целого и вещественного числа дает в результате вещественное число **float**.  
**float - int ---> float**

## Очередность выполнения операций

Порядок вычисления зависит от приоритета операций (order of operations). С математическими операторами Python следует математическому правилу.  
В таблице выше операторы отсортированы по приоритету выполнения - от максимального к минимальному (умножение и деление - операции одного приоритета).   
Операции одного приоритета выполняются в Python слева-направо (исключение - возведение в степень).

На практике лучше использовать скобки для группировки *операторов* (`+`, `-`, `/`, `*` и т.д.)  и *операндов*, чтобы в явном виде указать порядок вычисления выражений. Заодно это облегчит чтение программы.

### Пример

Рассчитаем структуру продаж бензина АИ-92, результаты запишем в переменные `percent_yanos` и `percent_mnpz` в зависимости от базиса отгрузки.

image.png

```python
percent_yanos = 840 / 840 + 600 * 100
print(percent_yanos)
```

```python
percent_mnpz = 600 / 840 + 600 * 100
print(percent_mnpz)
```

Первой выполнятся операции деления и умножения, затем сложения. Если оставить код в таком виде - расчет будет некорректен. 

Необходимо отрегулировать формулу и определить корректный порядок вычисления.

```python
percent_yanos = 840 / (840 + 600) * 100
print(percent_yanos)
```

```python
percent_mnpz = 600 / (840 + 600) * 100
print(percent_mnpz)
```

## Округление round()

В Python есть несколько встроенных функций, которые вы можете использовать для работы с числами.   
Часто требуется `round()` для округления чисел до некоторого количества знаков после запятой.   

Синтаксически функция `round()` вызывается двумя способами.

**round(x)** — это округление числа до целого, которое расположено ближе всего. Если дробная часть равна 0,5, то округляют до ближайшего целого значения.  
**round(x, n)** — данные х округляют до n знаков после точки. Если округление проходит до сотых, то n равен "2", если до тысячных — "3" и т.д.

### Пример

Округлим полученные доли отгрузок по заводам.

```python
# доля ЯНОС в структуре отгрузок
percent_yanos = 840 / (840 + 600) * 100
percent_yanos
```

```python
# по умолчанию округление до целого значения
percent_yanos = round(percent_yanos)
print(percent_yanos)
```

```python
# доля МНПЗ в структуре отгрузок
percent_mnpz = 600 / (840 + 600) * 100
percent_mnpz
```

```python
# округление до одного знака после запятой
percent_mnpz = round(percent_mnpz, 1)
print(percent_mnpz)
```

## Сохраняющее присваивание

Операторы этой группы используются для сохранения значения в переменной.

### Пример

Определим общее количество отгруженных товаров и запишем результат в перменную `total_sales`.

image.png

```python
# сумма отгрузок ЯНОС и МНПЗ
total_sales = 840 + 600 

# выведем на экран
print(total_sales)
```

Данные об отгрузках с биржи обновились, добавилась новая информация - актуализируем `total_sales`.  

*В примере здесь и выше исходные данные представлены в таблице для компактности, но работаем мы не с табличным форматом, как отдельным типом данных, а с отдельными числами. Оптимальные методы работы с табличными данными вы будете рассматривать при изучении библиотеки Pandas.*

image.png

```python
total_sales
```

```python
# сумма отгрузок ЯНОС, МНПЗ, ОНПЗ
total_sales = total_sales + 1320
print(total_sales)
```

Выражение `total_sales + 1320` можно записать компактнее, используя оператор присваивания `+=`

```python
# сумма отгрузок ЯНОС, МНПЗ
total_sales = 840 + 600 

# сумма отгрузок ЯНОС, МНПЗ, ОНПЗ
total_sales += 1320
print(total_sales)
```

Аналогичным образом работают операторы присваивания `-=`, `*=`, `/=`.
