## Именование переменных

**Синтаксические правила** для именования переменных в Python:  

- Имя переменной должно начинаться с буквы или символа подчеркивания.  
- Имя переменной не может начинаться с числа.  
- Имя переменной может содержать только буквенно-цифровые символы и символы подчеркивания (A-z, 0-9 и _).  
- Имена переменных чувствительны к регистру (price, Price и PRICE - это три разные переменные).  
- Имя переменной не должно совпадать с зарезервированным ключевым словом (print, def, return и т.д.).

Зарезервированных слов не так много - можно запомнить, но не обязательно. Вы всегда можете получить его, используя следующий код.

```python
import keyword
print(keyword.kwlist)
```

**Стилистические правила** для именования переменных в Python:  
- Имя должно отражать смысл переменной, легко читаться, не требовать дополнительных комментариев.  
- Преимущественно используется нотация **under_score** (все слова начинаются со строчной (маленькой) буквы, а в качестве разделителя между словами используется символ нижнего подчеркивания "_").  
- Используйте английские слова, не применяйте транслитерацию (русские слова латиницей).

## Преобразование типов

Для преобразования типов Python предоставляет ряд встроенных функций:  
`int()`: преобразует значение в целое число;  
`float()`: преобразует значение в число с плавающей точкой;  
`str()`: преобразует значение в строку.  

### Преобразование в целое число int()

```python
# цена бензина за 1 тонну
oil_price_per_tone = 37740     

# объем продаж за 1 день
sales_per_day = 565.5    

# выручка за 1 день
revenue_per_day = oil_price_per_tone * sales_per_day       
```

```python
print(revenue_per_day)
```

Переменная `revenue_per_day` фактически является целым числом, дробная часть отсутствует. В этой ситуации логично привести значение и тип переменной в соответствие.

```python
# преобразовали тип и присвоили его переменной
revenue_per_day = int(revenue_per_day) 

# вывод значения
print(revenue_per_day)                                  
```

```python
# вывод текущего типа переменной
print(type(revenue_per_day))                               
```

Когда переменная хранит вещественное число с ненулевой дробной частью, функция int() просто отбрасывает дробную часть.  
Положительные числа округляются вниз, а отрицательные — вверх.

```python
# объем продаж за 1 день
sales_per_day = 565.5 

# преобразовали тип переменной
print(int(sales_per_day))                                   
```

В ряде случаев возникает необходимость преобразования логического типа bool в целочисленный, в этом случае int() преобразует логическое значение в 1 или 0.

```python
# торгуется на бирже
traded_on_stock = True 

# преобразовали тип переменной
print(int(traded_on_stock))                                 
```

### Преобразование в число с плавающей запятой float()

Функция `float()` работает аналогично `int()`.

```python
# цена бензина за 1 тонну
oil_price_per_tone = 37740  
```

```python
# текущий тип данных - int()
print(type(oil_price_per_tone))
```

```python
# преобразовываем тип данных
oil_price_per_tone = float(oil_price_per_tone)
print(oil_price_per_tone)
```

```python
# выводим новый тип данных
print(type(oil_price_per_tone))
```

### Преобразование в строку str()

Функция `str()` может быть актуальна, например, при добавлении к строке значения другого типа.

```python
# наименование товара
item = 'Бензин (АИ-92-К5)'   

# цена бензина за 1 тонну 
oil_price_per_tone = 37740                                  
```

Без преобразования

```python
# выведем на экран фразу Рыночная цена Бензин (АИ-92-К5): 37740
trade_result = 'Рыночная цена ' + item + ':' + oil_price_per_tone
print(trade_result)
```

Без преобразования `oil_price_per_tone` - ошибка.

```python
trade_result = 'Рыночная цена ' + item + ':' + str(oil_price_per_tone)
print(trade_result)
```

## Итоги
1. Переменная – это ссылка на область памяти. После того как вы дали имя области, появляется возможность обращаться к данным, которые в ней хранятся.  
2. Чтобы создать переменную, нужно просто  написать имя переменной, поставить = (это оператор присваивания) и по правую сторону от знака равенства ввести значение, которое вы хотите присвоить переменной.
3. При именовании переменных необходимо придерживаться определенных синтаксических и стилистических правил.   
*Синтаксические правила*: имя переменной должно начинаться с буквы или символа подчеркивания, оно не может начинаться с числа, может содержать только буквенно-цифровые символы и символы подчеркивания (A-z, 0-9 и `_`), чувствительно к регистру, не должно совпадать с зарезервированным ключевым словом (print, def, return и т.д.).  
*Стилистические правила*: имя должно отражать смысл переменной, легко читаться, не требовать дополнительных комментариев, написано в нотации under_score с использованием английского языка.  
4. Переменная может содержать разные типы даных - числа, строки, списки, словари и т.д. Тип данных можно определить с помощью функции `type()`.  
5. Для преобразования типов переменных в Python существует ряд встроенных функций, таких как `int()`, `float()`, `str()`.
